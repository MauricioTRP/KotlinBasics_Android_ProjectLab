## Basic Operators

Kotlin operators work exactly like most C-family languages, with a few extras worth noting.

### Mathematical operators
| Operator | Meaning | Example |
|---|---|---|
| `+` | Addition | `3 + 2 = 5` |
| `-` | Subtraction | `3 - 2 = 1` |
| `*` | Multiplication | `3 * 2 = 6` |
| `/` | Division (integer if both operands are Int) | `5 / 2 = 2` |
| `%` | Modulo (remainder) | `5 % 2 = 1` |

> **Watch out:** Integer division truncates toward zero. `5 / 2` is `2`, not `2.5`. To get a decimal result, at least one operand must be a `Double` or `Float`.

### Comparison operators
Return a `Boolean`. Used in conditions and `when` expressions.
`==`, `!=`, `<`, `>`, `<=`, `>=`

### Logical operators
Combine boolean expressions:
- `&&` — AND (both must be true)
- `||` — OR (at least one must be true)
- `!` — NOT (negates the value)

### Assignment operators
`+=`, `-=`, `*=`, `/=`, `%=` are shorthand for `x = x <op> value`.

### Equality vs. reference equality
- `==` checks **structural equality** (calls `equals()` under the hood) — use this for comparing values.
- `===` checks **reference equality** (same object in memory) — rarely needed in day-to-day Kotlin.

````kotlin
// val "Constant"
val num1 = 5 // Inference of type by compiler
val num2: Int = 10 // Explicit type declaration

println("num1 + num2 = ${num1 + num2}") // String interpolation

println("num1 == num2 => ${num1 == num2}")
````

````kotlin
val num3 = num1 / num2

println("num1 class: ${num1.javaClass.name}")
println("num3 Class: ${num3.javaClass}")

println("num3 = ${num3}")
````

## Casting (Type Conversion)

In Kotlin, **numeric types are never implicitly converted** — there is no silent widening like in Java.
You must convert explicitly using the `toX()` methods.

```
Int  ──toDouble()──►  Double
Int  ──toLong()────►  Long
Double  ──toInt()──►  Int   (truncates the decimal part)
```

### Why explicit-only?

Implicit conversion is a common source of bugs. For example, in languages that allow it, dividing two integers and expecting a decimal result silently gives you the wrong answer. Kotlin makes you state your intent clearly.

### Safe casting with `as?`

When casting object types (not primitives), use `as?` instead of `as` to avoid a `ClassCastException`:
```kotlin
val obj: Any = "hello"
val str: String? = obj as? String   // returns null if the cast fails, not an exception
```

````kotlin
val num4: Double = num1.toDouble() / num2

println("num4 = $num4")
println("num4 class: ${num4.javaClass}")
````

## Numbers Are Objects

In Kotlin, **every number is an object** — `Int`, `Double`, `Long`, etc. are all full types with methods.
This means you can call functions on them directly, which makes the language uniform and expressive.

Under the hood, the Kotlin compiler is smart: when possible, it compiles `Int` to a primitive JVM `int` (for performance). The object wrapper only appears when required (e.g., generics, nullable `Int?`).

### Operator overloading
Kotlin's arithmetic operators (`+`, `-`, `*`, `/`) are actually method calls:

| Operator | Method equivalent |
|---|---|
| `a + b` | `a.plus(b)` |
| `a - b` | `a.minus(b)` |
| `a * b` | `a.times(b)` |
| `a / b` | `a.div(b)` |
| `a % b` | `a.rem(b)` |

This means you can define these operators on your own classes too — a powerful feature called **operator overloading**.

````kotlin
val num5 = num1.times(num2) // num1 * num2

println("num5 = ${num5}")
````

## Underscores for Readability

Kotlin lets you place underscores **anywhere inside a numeric literal** as a visual separator.
The compiler ignores them completely — they exist only for the human reader.

This works for all numeric bases:
```kotlin
val million    = 1_000_000        // decimal
val bytes      = 0b1101_0010      // binary
val hexColor   = 0xFF_EC_D8_A0    // hexadecimal
val creditCard = 1234_5678_9012_3456L
```

Use them whenever grouping digits in threes (or fours for binary/hex) makes the number easier to read at a glance.

````kotlin
val oneMillion = 1_000_000_000
println("oneMillion = $oneMillion")
val bytes = 0b11010010_10010010
val hex = 0xFF_00_A0
````

## Strings

A `String` in Kotlin is an **immutable** sequence of characters. Once created, its content cannot change.
Every "modification" (like `uppercase()` or `trim()`) returns a **new** String.

### Two kinds of string literals

**Escaped strings** — use backslash sequences for special characters:
```kotlin
val escaped = "Line 1\nLine 2\tTabbed"
// \n = newline, \t = tab, \\ = backslash, \" = quote
```

**Raw strings** (triple-quoted) — contain **any** character literally, including newlines and quotes:
```kotlin
val json = """
    {
        "name": "Kotlin",
        "version": 2.0
    }
""".trimIndent()   // trimIndent() removes the common leading whitespace
```

Raw strings are ideal for JSON, SQL queries, regex patterns, or any multi-line text where escaping would be painful.

### Strings are `CharSequence`
You can index into a string (`str[0]`), iterate over its characters, and use all standard collection-style functions like `filter`, `map`, and `take`.

````kotlin
val aloha = "Aloha tod@s"
val aloha2 = "Aloha tod@s\n\t"

val bigText = """
    This is a big text
    that can contain newlines and "quotes" without escaping
"""

val jsonExampleText = """
{
    "key": "value
}
"""
````

## String Concatenation

You can join strings with the `+` operator. Kotlin also automatically converts the right-hand side to a String when the left side is a String:

```kotlin
val greeting = "Hello" + " " + "World"   // "Hello World"
val age = 30
val msg = "Age: " + age                   // age.toString() is called automatically
```

### Prefer string templates over `+`

Concatenation with `+` creates intermediate String objects at each step — fine for two or three pieces, but inefficient in a loop or with many parts. More importantly, templates are simply **easier to read**.

```kotlin
// Hard to read with concatenation
val log = "[" + level + "] " + timestamp + " - " + message

// Much cleaner with a template
val log = "[$level] $timestamp - $message"
```

As a rule: use `+` only when you're combining two already-computed strings. Otherwise, reach for a template.

````kotlin
val hello = "Hello"
val name = "Juan"

println(hello + " " + name)
````

## String Templates

String templates let you **embed values and expressions directly inside a string literal**, avoiding manual concatenation.

### Simple variable: `$variable`
Use a bare `$` when inserting a single variable name — no braces needed:
```kotlin
val name = "Ana"
println("Hello, $name")   // Hello, Ana
```

### Expression: `${expression}`
Use curly braces when you need to call a method, do math, or access a property:
```kotlin
println("Hello, ${name.uppercase()}")        // Hello, ANA
println("2 + 2 = ${2 + 2}")                  // 2 + 2 = 4
println("Items: ${cart.size} in your cart")
```

### Why this matters
- No string-to-type conversion to write manually — the template calls `.toString()` for you.
- The expression is evaluated at the point of the string, so it always reflects the current value.
- Works inside triple-quoted (raw) strings too, giving you readable multi-line templates for things like email bodies or SQL queries.

````kotlin
println("Hello, $name")

// curly braces usage
println("Hello, ${name.lowercase()}")
````

## Constants and Variables

Kotlin draws a hard line between things that change and things that don't.

### `val` — read-only (prefer this by default)
`val` declares a reference that is assigned **once** and never reassigned. It is the Kotlin equivalent of `final` in Java.

```kotlin
val pi = 3.14159
// pi = 3.0  ← compile error: Val cannot be reassigned
```

> **Important:** `val` means the **reference** cannot be reassigned, not that the object itself is immutable. A `val list = mutableListOf(1, 2, 3)` still lets you call `list.add(4)` — you just can't point `list` at a different list.

### `var` — mutable
`var` declares a variable whose value can be changed at any time.

```kotlin
var score = 0
score += 10   // fine
score = -1    // also fine
```

### When to use which?
Start with `val`. Change to `var` only when you have a concrete reason to reassign.
- **`val`** makes code easier to reason about — you know the value never changes after initialization.
- **`val`** is safer in concurrent code — immutable references can't cause data races.
- Excessive `var` usage is often a sign that logic should be restructured.

````kotlin
val birthAge = 1989
var weigth = 90

//birthAge++ reassign error.
weigth++
````

## Control Flow

Kotlin's control flow constructs are similar to other languages, but with one important distinction: **`if` and `when` are expressions** — they return a value and can be used anywhere a value is expected.

### `if / else if / else`
Works as a statement (like Java) or as an expression:
```kotlin
// as a statement
if (x > 0) println("positive")

// as an expression — replaces the ternary operator (?:) from Java
val label = if (x > 0) "positive" else "non-positive"
```
Kotlin has **no ternary operator** (`? :`). The `if` expression fills that role.

### `when`
`when` is Kotlin's replacement for `switch`. It is far more powerful:
- Matches against values, ranges, types, and arbitrary conditions.
- Each branch is an expression — no `break` statements needed, no fall-through.
- When used as an expression, the compiler enforces that all cases are covered (exhaustive).

```kotlin
when (x) {
    0          -> println("zero")
    in 1..9    -> println("single digit")
    is String  -> println("it's a string")
    else       -> println("something else")
}
```

````kotlin
if (birthAge >= 18) {
    println("You have to work")
} else if (birthAge >= 65) {
    println("You can retire")
} else {
    println("You don't have to work")
}
````

````kotlin
val canDrink = if (birthAge >= 18) "Fondo fondo" else "please not"

println(canDrink)
````

## Ranges

A **range** represents a contiguous sequence of values and is created with the `..` operator.
Ranges work with numbers and characters, and integrate directly with `in`, `for`, and `when`.

### Types of ranges
| Expression | Meaning | Includes end? |
|---|---|---|
| `1..10` | 1 to 10 | Yes |
| `1 until 10` | 1 to 9 | No (exclusive end) |
| `10 downTo 1` | 10 to 1 (descending) | Yes |
| `1..10 step 2` | 1, 3, 5, 7, 9 | Yes, with step |

### Membership check with `in`
`in` tests whether a value falls within a range — more readable than `x >= 1 && x <= 10`:
```kotlin
if (score in 0..100) println("valid score")
```

### Using ranges in `when`
Ranges as `when` branches replace long chains of `if / else if`:
```kotlin
when (httpStatus) {
    in 200..299 -> "Success"
    in 400..499 -> "Client error"
    in 500..599 -> "Server error"
    else        -> "Other"
}
```

````kotlin
var responseCode = 201 //Created

if (responseCode in 200..299)
    println("The response code $responseCode is successful")
````

````kotlin
responseCode = 404

when (responseCode) {
    in 100..199 -> println("Informational response")
    in 200..299 -> println("Successful response")
    in 300..399 -> println("Redirection response")
    in 400..499 -> println("Client error response")
    in 500..599 -> println("Server error response")
    else -> println("Invalid response")
}
````

## Loops

Kotlin has three looping constructs plus a utility function. Each has a clear use case.

### `for` — iterating over a range or collection
```kotlin
for (i in 1..5) { ... }               // range
for (item in list) { ... }            // collection
for ((index, item) in list.withIndex()) { ... }  // with index
```

### `while` — loop while a condition is true
Use when the number of iterations is not known upfront:
```kotlin
while (queue.isNotEmpty()) {
    process(queue.poll())
}
```

### `do-while` — guaranteed first execution
The body runs **at least once** before the condition is checked:
```kotlin
do {
    val input = readLine()
} while (input != "quit")
```

### `repeat(n)` — simple counted repetition
A cleaner alternative to `for (i in 0 until n)` when you don't need the index:
```kotlin
repeat(3) { println("retry") }
```
Inside the lambda, `it` holds the current iteration count (0-based) if you do need it.

### `break` and `continue`
Work as in other languages. Kotlin also supports **labeled breaks** (`break@outerLoop`) to exit nested loops cleanly.

````kotlin
for (i in 1..5) {
    println("i = $i")
}
````

````kotlin
for (i in 5 downTo 1) {
    println("i = $i")
}
````

````kotlin
for (i in 1..10 step 2) {
    println("i = $i")
}
````

````kotlin
val pets = listOf<String>("Llanka", "Mike", "Yuyo")

for (pet in pets) {
    println(pet)
}

// With indexes
for ((index, name) in pets.withIndex()) {
    println("$index: $name")
}
````

````kotlin
// range of ascii values

for (char in 'a'..'z') {
    println(char)
}
````

````kotlin
var count = 0

while (count < 10) {
    println("counting: $count")

    count++
}
````

````kotlin
repeat(count) {
    println("Counting $it")
}
````

## Lists

A `List` is an **ordered collection** that allows duplicate elements and provides index-based access.

### Immutable vs. mutable
Kotlin separates the **read interface** from the **write interface**:

| Function | Type | Can add/remove? |
|---|---|---|
| `listOf(...)` | `List<T>` | No |
| `mutableListOf(...)` | `MutableList<T>` | Yes |

`List<T>` only exposes read operations (`get`, `size`, `contains`, iteration).
`MutableList<T>` adds `add`, `remove`, `set`, and `clear`.

> **Prefer `List<T>` in function signatures.** If a caller passes a `MutableList`, your function receives it as read-only — the list cannot be accidentally mutated inside the function.

### Key operations
```kotlin
val items = listOf("a", "b", "c")
items[0]            // "a" — index access
items.size          // 3
items.contains("b") // true
items.first()       // "a"
items.last()        // "c"
items.filter { it != "b" }   // ["a", "c"]  (returns a new list)
```

````kotlin
data class Instrument(val name: String, val type: String)

// Immutable
val instruments = listOf<Instrument>(
    Instrument(name = "Trumpet", type = "Brass")
)

val mutableList = mutableListOf<Instrument>()

for (instrument in instruments) {
    println(instrument.name)
}
````

````kotlin
//instruments.add(Instrument(name = "Trombone", type = "Brass")) // Error

mutableList.add(Instrument(name = "Trombone", type = "Brass"))
mutableList.remove(Instrument(name = "Trombone", type = "Brass"))
````

## Arrays

An `Array<T>` is a **fixed-size**, **mutable** container. Its size is set at creation and cannot change.

### Arrays vs. Lists

| | `Array<T>` | `List<T>` / `MutableList<T>` |
|---|---|---|
| Size | Fixed | Dynamic (`MutableList` can grow) |
| Elements | Always mutable | Depends on type |
| Interop | Direct JVM array (fast) | Kotlin collection interface |
| Common use | Low-level / interop | Day-to-day Kotlin code |

**Prefer `List` in most Kotlin code.** Use `Array` when you need a JVM primitive array for performance or when interoperating with a Java API that expects one.

### Primitive arrays
For performance-critical code, use typed arrays that avoid boxing:
```kotlin
val ints   = IntArray(5)          // equivalent to int[] in Java
val floats = FloatArray(10) { it * 0.1f }
```

### Creation
```kotlin
val arr = arrayOf(1, 2, 3)       // Array<Int> (boxed)
val arr2 = Array(5) { it * 2 }   // [0, 2, 4, 6, 8] via lambda
arr[0] = 99                       // mutation is allowed
```

````kotlin
sealed class Note {
    data class C(val octave: Int) : Note()
    data class D(val octave: Int) : Note()
    data class E(val octave: Int) : Note()
    data class F(val octave: Int) : Note()
    data class G(val octave: Int) : Note()
    data class A(val octave: Int) : Note()
    data class B(val octave: Int) : Note()
}

var octave4 = arrayOf <Note>(
    Note.C(4),
    Note.D(4),
    Note.E(4),
    Note.F(4),
    Note.G(4),
    Note.A(4),
    Note.B(4)
)

octave4[0] = Note.C(5)
````

# Null Safety

`NullPointerException` (NPE) is famously called the **"Billion Dollar Mistake"** — a term coined by Tony Hoare, who invented the null reference in 1965 and later regretted it. In Java and many other languages, any reference can be `null`, and the compiler doesn't warn you. You only find out at runtime when the app crashes.

**Kotlin eliminates this class of bugs at compile time** by building nullability into the type system.

### Nullable vs. non-nullable types
Every type in Kotlin is **non-nullable by default**:
```kotlin
var name: String = "Ana"
name = null   // ← compile error! String cannot hold null
```

To allow null, append `?` to the type:
```kotlin
var name: String? = "Ana"
name = null   // fine — String? can hold null
```

### Safe call operator `?.`
Access a property or call a method on a nullable type without risking an NPE.
If the value is `null`, the entire expression short-circuits and returns `null` instead of crashing:
```kotlin
val length = name?.length   // returns Int? — null if name is null
```

You can chain safe calls:
```kotlin
val city = user?.address?.city   // null if user or address is null
```

````kotlin
// Cannot assign Null to a type
// val myNumber : Int = null // Error null assignment

var numberOfPeople : Int? = null

numberOfPeople?.plus(1)
````

## Not-Null Assertion (`!!`)

The `!!` operator tells the compiler: _"I am certain this value is not null right now — skip the null check."_

If you are wrong and the value **is** null at runtime, Kotlin throws a `NullPointerException` immediately at that line, with a clear stack trace pointing to the exact spot.

```kotlin
val length = name!!.length   // NPE here if name is null
```

### When to use `!!`
Use it **only** when you have external knowledge the compiler can't see — for example, a value that was set by a framework before your code runs, or a `lateinit var` that you know is initialized. It is an explicit contract between you and the compiler.

### When NOT to use `!!`
Do not use `!!` to silence compiler warnings. That defeats the purpose of null safety.

| Situation | Better alternative |
|---|---|
| Need a default when null | Elvis operator `?: defaultValue` |
| Want to skip if null | Safe call `?.let { ... }` |
| Need to check before using | `if (value != null)` smart cast |
| Just want to compile | Fix the nullability properly |

> A codebase full of `!!` provides no more safety guarantees than Java. Treat each `!!` as a code smell that needs a comment explaining why it is safe.

````kotlin
numberOfPeople = null

println(numberOfPeople!!.times(2))
````

## Elvis Operator (`?:`)

The `?:` operator is named after Elvis Presley — tilt it 90° and it looks like his iconic hair and eyes. 🕺

It provides a **default value** when a nullable expression evaluates to `null`:

```kotlin
val length = name?.length ?: 0
// If name is null  → name?.length is null → result is 0
// If name is "hi"  → name?.length is 2   → result is 2
```

### Chaining with safe calls
The Elvis operator pairs naturally with `?.` to form a safe, concise idiom:
```kotlin
val display = user?.profile?.displayName ?: "Anonymous"
```

### Elvis with `return` or `throw`
The right-hand side of `?:` can be any expression — including `return` or `throw`.
This is commonly used for early-exit guards at the top of a function:
```kotlin
fun processOrder(order: Order?) {
    val id = order?.id ?: return          // exit early if null
    val price = order.price ?: throw IllegalStateException("No price set")
    // ... rest of the function with non-null values
}
```

This pattern replaces the classic Java null-check boilerplate with a single, readable line.

````kotlin
println(numberOfPeople ?: 0)
````
