## Basic operators

- Mathematical operators: `+`, `-`, `*`, `/`, `%`
- Comparison operators: `==`, `!=`, `<`, `>`, `<=`, `>=`
- Logical operators: `&&`, `||`, `!`
- Assignment operators: `=`, `+=`, `-=`, `*=`, `/=`, `%=`
- Equality operators: `===`, `!==` (reference equality)

```kotlin
// val "Constant"
val num1 = 5 // Inference of type by compiler
val num2: Int = 10 // Explicit type declaration

println("num1 + num2 = ${num1 + num2}") // String interpolation

println("num1 == num2 => ${num1 == num2}")
```

```kotlin
val num3 = num1 / num2

println("num1 class: ${num1.javaClass.name}")
println("num3 Class: ${num3.javaClass}")

println("num3 = ${num3}")
```

## Casting

reinterpretamos un tipo respecto a otro tipo

```kotlin
val num4: Double = num1.toDouble() / num2

println("num4 = $num4")
println("num4 class: ${num4.javaClass}")
```

## Numbers are objects

you can send "messages" (methods) to numbers, just like any other object in Kotlin

```kotlin
val num5 = num1.times(num2) // num1 * num2

println("num5 = ${num5}")
```

## Strings

- String literals are enclosed in double quotes (`"`)

```kotlin
val aloha = "Aloha tod@s"
val aloha2 = "Aloha tod@s\n\t"

val bigText = """
    This is a big text
    that can contain newlines and "quotes" without escaping
"""

val jsonExampleText = """
{
    "key": "value
}
"""
```

## Contatenation

- You can concatenate strings using the `+` operator or string templates

```kotlin
val hello = "Hello"
val name = "Juan"

println(hello + " " + name)
```

## String templates

- String templates allow you to embed expressions inside string literals using `${expression}`
- You can also use simple variable names with sign $, (without operations)

```kotlin
println("Hello, $name")

// curly braces usage
println("Hello, ${name.lowercase()}")
```

## Control Flow

```kotlin
if (birthAge >= 18) {
    println("You have to work")
} else if (birthAge >= 65) {
    println("You can retire")
} else {
    println("You don't have to work")
}
```

## Ranges

- Ranges are defined using the `..` operator

```kotlin
var responseCode = 201 //Created

if (responseCode in 200..299)
    println("The response code $responseCode is successful")
```

```kotlin
responseCode = 404

when (responseCode) {
    in 100..199 -> println("Informational response")
    in 200..299 -> println("Successful response")
    in 300..399 -> println("Redirection response")
    in 400..499 -> println("Client error response")
    in 500..599 -> println("Server error response")
    else -> println("Invalid response")
}
```

## Loops

- For loops
- While loops
- Do-while loops

```kotlin
for (i in 1..5) {
    println("i = $i")
}
```

```kotlin
for (i in 5 downTo 1) {
    println("i = $i")
}
```

```kotlin
for (i in 1..10 step 2) {
    println("i = $i")
}
```

```kotlin
val pets = listOf<String>("Llanka", "Mike", "Yuyo")

for (pet in pets) {
    println(pet)
}

// With indexes
for ((index, name) in pets.withIndex()) {
    println("$index: $name")
}
```

```kotlin
// range of ascii values

for (char in 'a'..'z') {
    println(char)
}
```

```kotlin
var count = 0

while (count < 10) {
    println("counting: $count")

    count++
}
```

```kotlin
repeat(count) {
    println("Counting $it")
}
```

## Lists

- Lists are ordered collections of elements
- They can be mutable or immutable

```kotlin
data class Instrument(val name: String, val type: String)

// Immutable
val instruments = listOf<Instrument>(
    Instrument(name = "Trumpet", type = "Brass")
)

val mutableList = mutableListOf<Instrument>()

for (instrument in instruments) {
    println(instrument.name)
}
```

```kotlin
//instruments.add(Instrument(name = "Trombone", type = "Brass")) // Error

mutableList.add(Instrument(name = "Trombone", type = "Brass"))
mutableList.remove(Instrument(name = "Trombone", type = "Brass"))
```

## Arrays

- Arrays are fixed-size collections of elements
- They can be of primitive types or objects

```kotlin
sealed class Note {
    data class C(val octave: Int) : Note()
    data class D(val octave: Int) : Note()
    data class E(val octave: Int) : Note()
    data class F(val octave: Int) : Note()
    data class G(val octave: Int) : Note()
    data class A(val octave: Int) : Note()
    data class B(val octave: Int) : Note()
}

var octave4 = arrayOf <Note>(
    Note.C(4),
    Note.D(4),
    Note.E(4),
    Note.F(4),
    Note.G(4),
    Note.A(4),
    Note.B(4)
)

octave4[0] = Note.C(5)
```

# Null Safety

- Kotlin has built-in null safety features to prevent null pointer exceptions

```kotlin
// Cannot assign Null to a type
// val myNumber : Int = null // Error null assignment

var numberOfPeople : Int? = null

numberOfPeople?.plus(1)
```

## Not null assertion

!!: operator assures the compiler that the value is not null, and if it is null, it will throw a NullPointerException

```kotlin
numberOfPeople = null

println(numberOfPeople!!.times(2))
```

## ?: Elvis Operator

- The Elvis operator (`?:`) is used to provide a default value when an expression is null

```kotlin
println(numberOfPeople ?: 0)
```
