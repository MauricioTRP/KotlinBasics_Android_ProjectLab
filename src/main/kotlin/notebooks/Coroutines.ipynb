# Coroutines

With coroutines we allow our code to perform multiple tasks at once. This concept is known as concurrency.
A coroutine is a suspendable computation that lets you write asynchronous, non-blocking code in a sequential manner. Coroutines are a powerful tool for managing concurrency in Kotlin, allowing you to write code that is more efficient and easier to read.

On JVM and Kotlin/Native all concurrent code runs on **threads** managed by OS.

![Parallelism and Concurrency Diagram](https://kotlinlang.org/docs/images/parallelism-and-concurrency.svg)

A powerful feature of coroutines is that they can be suspended and resumed without blocking the underlying thread. This allows you to write code that is more efficient and responsive, as it can perform other tasks while waiting for a long-running operation to complete. (e.g., network requests, database operations, etc.)

```kotlin
import kotlinx.coroutines.runBlocking

fun main() = runBlocking {
    println("Hello, World!")
}
```
